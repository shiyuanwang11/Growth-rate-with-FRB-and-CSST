```python
from getdist import loadMCSamples,plots
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

def rgba(cc):
    return tuple(cc/255.)
c1 = rgba(np.array([41.,157.,143.,255.]))
c2 = rgba(np.array([233.,196.,106.,255.]))
c3 = rgba(np.array([216.,118.,89.,255.]))

samp1 = loadMCSamples('/home/wangsy2/software/MGCobaya/chains_LCDM_t3/mugamma')
p1 = samp1.getParams()
samp1.addDerived(p1.omega_de * p1.E11, name='mu1', label=r'\mu_0-1')
samp1.addDerived(p1.omega_de * p1.E22, name='eta1', label=r'\eta_0-1')

samp2 = loadMCSamples('/home/wangsy2/software/MGCobaya/chains_LCDM_t3/mugamma.post.ksz')
p2 = samp2.getParams() 
samp2.addDerived(p2.omega_de * p2.E11, name='mu1', label=r'\mu_0-1')
samp2.addDerived(p2.omega_de * p2.E22, name='eta1', label=r'\eta_0-1')

samp3 = loadMCSamples('/home/wangsy2/software/MGCobaya/chains_LCDM_t3/mugamma.post.rsd')
p3 = samp3.getParams() 
samp3.addDerived(p3.omega_de * p3.E11, name='mu1', label=r'\mu_0-1')
samp3.addDerived(p3.omega_de * p3.E22, name='eta1', label=r'\eta_0-1')

g = plots.get_single_plotter()
g.add_x_marker(0)
g.add_y_marker(0)
g.plot_2d([samp1,samp3,samp2], 'eta1', 'mu1', filled=True, colors=[c1,c3,c2])
g.add_legend(['Planck','Planck+RSD','Planck+RSD+kSZ+FRB'])
for ax in g.subplots[:, :].flat:
    if ax is not None:  #  ax  None
        ax.minorticks_on()
        ax.tick_params(which='both', direction='in', top=True, right=True)
        ax.tick_params(which='minor', length=4)
        ax.tick_params(which='major', length=7)
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.8
g.settings.title_limit_fontsize = 14
g.export('fr_mueta_full.pdf')
```
